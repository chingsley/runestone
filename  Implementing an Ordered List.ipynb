```python
class Node:
    def __init__(self, initialdata):
        self.data = initialdata
        self.next = None
    
    def getData(self):
        return self.data
    
    def getNext(self):
        return self.next
    
    def setData(newdata):
        self.data = newdata
        
    def setNext(self, newnext):
        self.next = newnext
```
